```python
import numpy as np
import pandas as pd
import os
```

```python
def add_student():
    print("\n--- Enter Student Details ---")
    name = input("Student Name: ")
    maths = input("Maths score: ")
    python = input("Python score: ")
    physics = input("Physics score: ")
    oops = input("Oops score: ")
        
        # Check if file exists to decide if we need a header
    file_exists = os.path.isfile('grade.csv')
    
        # Save to CSV (using append mode 'a')
    with open('grade.csv', 'a') as f:
            # If file is new, write the header first
        if not file_exists:
            f.write("Name,Maths,Python,Physics,Oops\n")
            # Write the student data
        f.write(f"{name},{maths},{python},{physics},{oops}\n")
    print(f"Successfully added {name}!")

# 2. Add as many students as you want
while True:
    add_student()
    cont = input("Add another student? (y/n): ")
    if cont.lower() != 'y':
        break
```

```python
#for calculating scores
assignment_mean=np.mean(grades_arr,axis=0)
print(f"Average score: {assignment_mean}")
assignment_median=np.median(grades_arr,axis=0)
print(f"Median : {assignment_median}")
assignment_min=np.min(grades_arr,axis=0)
print(f"Minimum score: {assignment_min}")
assignment_max=np.max(grades_arr,axis=0)
print(f"Maximum score: {assignment_max}")
```

```python
student_average=np.mean(grades_arr,axis=1)
student_data['final_grade']=student_average

student_data.sort_values(by='final_grade',ascending=False,inplace=True)
print(student_data)
```
