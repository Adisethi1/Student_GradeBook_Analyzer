```python
import numpy as np
import pandas as pd
import os
```

```python
assignment_info={"Maths":[67,89,99,100],
                "Python":[57,93,89,91],
                "Physics":[70,89,96,99],
                "Oops":[80,79,100,96]}
students=["Bob","Charlie","Aditya","Bhumika"]

student_data =pd.DataFrame(assignment_info,index=students) 
grades_arr=student_data.values
```

```python
#for calculating scores
assignment_mean=np.mean(grades_arr,axis=0)
print(f"Average score: {assignment_mean}")
assignment_median=np.median(grades_arr,axis=0)
print(f"Median : {assignment_median}")
assignment_min=np.min(grades_arr,axis=0)
print(f"Minimum score: {assignment_min}")
assignment_max=np.max(grades_arr,axis=0)
print(f"Maximum score: {assignment_max}")
```

```python
student_average=np.mean(grades_arr,axis=1)
student_data['final_grade']=student_average

student_data.sort_values(by='final_grade',ascending=False,inplace=True)
print(student_data)
```
